```python
import os
from pydub import AudioSegment
```

```python
mp3_folder = "E:/Education/4 course 1 semester/Course project/Shards_prodject/Code/Task/Dataset/cv-corpus-23.0-2025-09-05/ru/clips"
wav_folder = "E:/Education/4 course 1 semester/Course project/Shards_prodject/Code/Task/Dataset/cv-corpus-23.0-2025-09-05/ru/clips_wav"
os.makedirs(wav_folder, exist_ok=True)
```

```python
for filename in os.listdir(mp3_folder):
    if not filename.lower().endswith(".mp3"):
        continue


    mp3_path = os.path.join(mp3_folder, filename)
    wav_filename = os.path.splitext(filename)[0] + ".wav"
    wav_path = os.path.join(wav_folder, wav_filename)


    audio = AudioSegment.from_mp3(mp3_path)
    audio = audio.set_channels(1)
    audio = audio.set_frame_rate(16000)
    audio.export(wav_path, format="wav")


print("Конвертация всех MP3 в WAV завершена!")
```
